```python
# =============================================================================
# CELL 5: DATA LOADING AND PREPARATION
# =============================================================================
# Loading the Adult Income dataset as a proxy for hiring data
# =============================================================================

display(HTML("""
<div class="section-header">
    Section 2: Data Loading and Preparation
</div>
"""))

display(HTML("""
<div class="warning-box">
    <strong>Important Data Disclosure</strong><br><br>
    <strong>Why we're using the Adult Income Dataset:</strong><br>
    Real AI hiring system data is proprietary and contains sensitive personal information. 
    For this educational audit, we use the UCI Adult Income dataset, which:
    <ul>
        <li>Contains similar demographic attributes (age, gender, race, education)</li>
        <li>Has a binary outcome (income >$50K / $50K) that can proxy hiring decisions</li>
        <li>Is widely used in fairness research for demonstrating bias detection methods</li>
    </ul>
    <br>
    <strong>Limitation:</strong> The Adult dataset is from the 1994 US Census and doesn't perfectly 
    represent South African demographics. In a real audit, you would use actual South African 
    hiring data with appropriate consent.
    <br><br>
    <strong>Disability Simulation:</strong> The original dataset lacks disability information. 
    We will <strong>simulate</strong> a disability column based on realistic prevalence rates to 
    demonstrate the methodology. This is clearly disclosed and labeled throughout.
</div>
"""))

print("Loading Adult Income Dataset...\n")

# Define column names
column_names = [
    'age', 'workclass', 'fnlwgt', 'education', 'education_num',
    'marital_status', 'occupation', 'relationship', 'race', 'sex',
    'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income'
]

# Load data
try:
    url_train = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
    df = pd.read_csv(url_train, names=column_names, na_values=' ?', skipinitialspace=True)
    print(f"Successfully loaded {len(df):,} records from UCI repository")
except Exception as e:
    print(f"Could not load from URL, using backup method: {e}")
    from sklearn.datasets import fetch_openml
    adult = fetch_openml('adult', version=2, as_frame=True)
    df = adult.frame
    df.columns = column_names[:len(df.columns)]

print(f"Dataset Shape: {df.shape[0]:,} rows x {df.shape[1]} columns")
print(f"Data Source: UCI Machine Learning Repository (1994 US Census)")

# =============================================================================
# DATA CLEANING
# =============================================================================

display(HTML("""<div class="subsection-header">Data Cleaning</div>"""))

print("Checking for missing values...\n")
missing = df.isnull().sum()
if missing.sum() > 0:
    print(f"Dropping {df.isnull().any(axis=1).sum():,} rows with missing values...")
    df = df.dropna()
else:
    print("No missing values found.")

print(f"Clean dataset shape: {df.shape[0]:,} rows x {df.shape[1]} columns")

# =============================================================================
# FEATURE ENGINEERING
# =============================================================================

display(HTML("""<div class="subsection-header">Feature Engineering & Protected Attributes</div>"""))

data = df.copy()

# 1. GENDER
print("1. Processing Gender attribute...")
data['gender'] = (data['sex'] == 'Male').astype(int)
gender_dist = data['sex'].value_counts()
print(f"   Male: {gender_dist.get('Male', 0):,} ({gender_dist.get('Male', 0)/len(data)*100:.1f}%)")
print(f"   Female: {gender_dist.get('Female', 0):,} ({gender_dist.get('Female', 0)/len(data)*100:.1f}%)")

# 2. RACE
print("\n2. Processing Race attribute...")
data['race_binary'] = (data['race'] == 'White').astype(int)
data['race_original'] = data['race']
race_dist = data['race'].value_counts()
for race, count in race_dist.items():
    print(f"   {race}: {count:,} ({count/len(data)*100:.1f}%)")

# 3. AGE
print("\n3. Processing Age attribute...")
data['age_group'] = pd.cut(data['age'], 
                           bins=[0, 25, 35, 45, 55, 100],
                           labels=['18-25', '26-35', '36-45', '46-55', '55+'])
data['age_binary'] = (data['age'] >= 40).astype(int)
print(f"   Under 40: {(data['age_binary'] == 0).sum():,} ({(data['age_binary'] == 0).sum()/len(data)*100:.1f}%)")
print(f"   40 and over: {(data['age_binary'] == 1).sum():,} ({(data['age_binary'] == 1).sum()/len(data)*100:.1f}%)")

# 4. DISABILITY (SIMULATED)
print("\n4. Simulating Disability attribute...")
print("   NOTE: Disability status is SIMULATED for demonstration purposes")
np.random.seed(42)
disability_base_rate = 0.10
age_factor = (data['age'] - 18) / (data['age'].max() - 18) * 0.10
disability_prob = disability_base_rate + age_factor
data['disability'] = np.random.binomial(1, disability_prob)
print(f"   Simulated prevalence: {data['disability'].mean()*100:.1f}%")

# 5. TARGET VARIABLE
print("\n5. Processing Target Variable...")
data['target'] = (data['income'] == '>50K').astype(int)
target_dist = data['target'].value_counts()
print(f"   Positive outcome (>$50K): {target_dist.get(1, 0):,} ({target_dist.get(1, 0)/len(data)*100:.1f}%)")
print(f"   Negative outcome ($50K): {target_dist.get(0, 0):,} ({target_dist.get(0, 0)/len(data)*100:.1f}%)")

# Save processed data
processed_data = data.copy()
feature_columns = ['age', 'education_num', 'hours_per_week', 'capital_gain', 
                   'capital_loss', 'gender', 'race_binary', 'disability']

display(HTML("""
<div class="success-box">
    <strong>Data Preparation Complete</strong><br><br>
    <table style="width: 100%; border-collapse: collapse;">
        <tr><td style="padding: 8px;"><strong>Total Records:</strong></td><td>""" + f"{len(data):,}" + """</td></tr>
        <tr><td style="padding: 8px;"><strong>Protected Attributes:</strong></td><td>Gender, Race, Age, Disability</td></tr>
        <tr><td style="padding: 8px;"><strong>Target Variable:</strong></td><td>Income > $50K (proxy for hiring)</td></tr>
    </table>
</div>
"""))

display(HTML('<div class="section-divider"></div>'))
```

```python
# =============================================================================
# CELL 8: FAIRNESS METRICS CALCULATION
# =============================================================================
# Calculating the three required fairness metrics using AIF360
# =============================================================================

display(HTML("""
<div class="section-header">
    Section 5: Fairness Metrics Calculation
</div>
"""))

display(HTML("""
<div class="legal-box">
    <div class="legal-box-title">Understanding Fairness Metrics in Legal Context</div>
    <p style="margin: 0; color: #374151; line-height: 1.7;">
        These metrics help us quantify whether the AI system treats protected groups fairly, 
        as required by Section 9 of the Constitution and Section 6 of the Employment Equity Act. 
        Each metric measures a different aspect of fairness.
    </p>
</div>
"""))

# =============================================================================
# METRIC DEFINITIONS DISPLAY
# =============================================================================

metric_definitions = """
<div style="display: grid; grid-template-columns: repeat(3, 1fr); gap: 20px; margin: 25px 0;">
    
    <div style="background: white; border-radius: 16px; padding: 25px; 
                box-shadow: 0 4px 20px rgba(0,0,0,0.08); border-top: 4px solid #8B5CF6;">
        <h4 style="color: #8B5CF6; margin: 0 0 15px 0;">Statistical Parity Difference</h4>
        <p style="color: #6B7280; font-size: 0.9em; margin: 0 0 15px 0;">
            Measures the difference in positive outcome rates between groups.
        </p>
        <div style="background: #F9FAFB; padding: 12px; border-radius: 8px; margin-bottom: 15px;">
            <code style="color: #6D28D9;">SPD = P(Y=1|Unpriv) - P(Y=1|Priv)</code>
        </div>
        <div style="font-size: 0.85em;">
            <strong>Ideal value:</strong> 0<br>
            <strong>Fair range:</strong> -0.1 to 0.1<br>
            <strong>Negative:</strong> Unprivileged disadvantaged
        </div>
    </div>
    
    <div style="background: white; border-radius: 16px; padding: 25px; 
                box-shadow: 0 4px 20px rgba(0,0,0,0.08); border-top: 4px solid #F472B6;">
        <h4 style="color: #F472B6; margin: 0 0 15px 0;">Disparate Impact Ratio</h4>
        <p style="color: #6B7280; font-size: 0.9em; margin: 0 0 15px 0;">
            Ratio of positive outcome rates (the "80% rule" in US law).
        </p>
        <div style="background: #F9FAFB; padding: 12px; border-radius: 8px; margin-bottom: 15px;">
            <code style="color: #6D28D9;">DIR = P(Y=1|Unpriv) / P(Y=1|Priv)</code>
        </div>
        <div style="font-size: 0.85em;">
            <strong>Ideal value:</strong> 1.0<br>
            <strong>Fair range:</strong> 0.8 to 1.25<br>
            <strong>Below 0.8:</strong> Potential discrimination
        </div>
    </div>
    
    <div style="background: white; border-radius: 16px; padding: 25px; 
                box-shadow: 0 4px 20px rgba(0,0,0,0.08); border-top: 4px solid #2DD4BF;">
        <h4 style="color: #0D9488; margin: 0 0 15px 0;">Equal Opportunity Difference</h4>
        <p style="color: #6B7280; font-size: 0.9em; margin: 0 0 15px 0;">
            Difference in true positive rates between groups.
        </p>
        <div style="background: #F9FAFB; padding: 12px; border-radius: 8px; margin-bottom: 15px;">
            <code style="color: #6D28D9;">EOD = TPR(Unpriv) - TPR(Priv)</code>
        </div>
        <div style="font-size: 0.85em;">
            <strong>Ideal value:</strong> 0<br>
            <strong>Fair range:</strong> -0.1 to 0.1<br>
            <strong>Negative:</strong> Qualified individuals missed
        </div>
    </div>
    
</div>
"""

display(HTML(metric_definitions))

# =============================================================================
# FAIRNESS CALCULATION FUNCTION
# =============================================================================

def calculate_fairness_metrics(y_true, y_pred, protected_attr, privileged_value=1):
    """
    Calculate all three fairness metrics for a given protected attribute.
    
    Parameters:
    -----------
    y_true : array - True labels
    y_pred : array - Predicted labels
    protected_attr : array - Protected attribute values
    privileged_value : int - Value indicating privileged group (default=1)
    
    Returns:
    --------
    dict : Dictionary containing all fairness metrics
    """
    # Separate groups
    priv_mask = protected_attr == privileged_value
    unpriv_mask = protected_attr != privileged_value
    
    # Positive outcome rates
    priv_positive_rate = y_pred[priv_mask].mean()
    unpriv_positive_rate = y_pred[unpriv_mask].mean()
    
    # 1. Statistical Parity Difference
    spd = unpriv_positive_rate - priv_positive_rate
    
    # 2. Disparate Impact Ratio
    if priv_positive_rate > 0:
        dir_ratio = unpriv_positive_rate / priv_positive_rate
    else:
        dir_ratio = np.nan
    
    # 3. Equal Opportunity Difference (True Positive Rates)
    # TPR = TP / (TP + FN) = TP / Actual Positives
    priv_actual_pos = y_true[priv_mask] == 1
    unpriv_actual_pos = y_true[unpriv_mask] == 1
    
    if priv_actual_pos.sum() > 0:
        priv_tpr = y_pred[priv_mask][priv_actual_pos].mean()
    else:
        priv_tpr = np.nan
        
    if unpriv_actual_pos.sum() > 0:
        unpriv_tpr = y_pred[unpriv_mask][unpriv_actual_pos].mean()
    else:
        unpriv_tpr = np.nan
    
    eod = unpriv_tpr - priv_tpr
    
    # Additional metrics for context
    priv_count = priv_mask.sum()
    unpriv_count = unpriv_mask.sum()
    
    return {
        'statistical_parity_difference': spd,
        'disparate_impact_ratio': dir_ratio,
        'equal_opportunity_difference': eod,
        'privileged_positive_rate': priv_positive_rate,
        'unprivileged_positive_rate': unpriv_positive_rate,
        'privileged_tpr': priv_tpr,
        'unprivileged_tpr': unpriv_tpr,
        'privileged_count': priv_count,
        'unprivileged_count': unpriv_count
    }

def assess_fairness(metrics):
    """
    Assess whether metrics meet fairness thresholds.
    
    Returns dict with pass/fail status for each metric.
    """
    assessments = {}
    
    # Statistical Parity Difference: Fair if between -0.1 and 0.1
    spd = metrics['statistical_parity_difference']
    assessments['spd_fair'] = -0.1 <= spd <= 0.1
    assessments['spd_status'] = 'PASS' if assessments['spd_fair'] else 'FAIL'
    
    # Disparate Impact Ratio: Fair if between 0.8 and 1.25
    dir_val = metrics['disparate_impact_ratio']
    assessments['dir_fair'] = 0.8 <= dir_val <= 1.25
    assessments['dir_status'] = 'PASS' if assessments['dir_fair'] else 'FAIL'
    
    # Equal Opportunity Difference: Fair if between -0.1 and 0.1
    eod = metrics['equal_opportunity_difference']
    assessments['eod_fair'] = -0.1 <= eod <= 0.1
    assessments['eod_status'] = 'PASS' if assessments['eod_fair'] else 'FAIL'
    
    # Overall assessment
    assessments['overall_fair'] = all([
        assessments['spd_fair'], 
        assessments['dir_fair'], 
        assessments['eod_fair']
    ])
    
    return assessments

# =============================================================================
# CALCULATE METRICS FOR ALL PROTECTED ATTRIBUTES
# =============================================================================

display(HTML("""<div class="subsection-header">Calculating Fairness Metrics by Protected Attribute</div>"""))

# Prepare test data arrays
y_test_arr = y_test.values
y_pred_arr = y_pred_test

# Store all results
fairness_results = {}

# 1. GENDER FAIRNESS
print("Calculating fairness metrics for GENDER...")
gender_metrics = calculate_fairness_metrics(
    y_test_arr, 
    y_pred_arr, 
    X_test['gender'].values,
    privileged_value=1  # Male = privileged
)
gender_assessment = assess_fairness(gender_metrics)
fairness_results['Gender'] = {**gender_metrics, **gender_assessment}
print(f"  Statistical Parity Difference: {gender_metrics['statistical_parity_difference']:.4f}")
print(f"  Disparate Impact Ratio: {gender_metrics['disparate_impact_ratio']:.4f}")
print(f"  Equal Opportunity Difference: {gender_metrics['equal_opportunity_difference']:.4f}")

# 2. RACE FAIRNESS
print("\nCalculating fairness metrics for RACE...")
race_metrics = calculate_fairness_metrics(
    y_test_arr,
    y_pred_arr,
    X_test['race_binary'].values,
    privileged_value=1  # White = privileged
)
race_assessment = assess_fairness(race_metrics)
fairness_results['Race'] = {**race_metrics, **race_assessment}
print(f"  Statistical Parity Difference: {race_metrics['statistical_parity_difference']:.4f}")
print(f"  Disparate Impact Ratio: {race_metrics['disparate_impact_ratio']:.4f}")
print(f"  Equal Opportunity Difference: {race_metrics['equal_opportunity_difference']:.4f}")

# 3. AGE FAIRNESS (need to calculate age_binary for test set)
print("\nCalculating fairness metrics for AGE...")
age_binary_test = (X_test['age'] >= 40).astype(int).values
age_metrics = calculate_fairness_metrics(
    y_test_arr,
    y_pred_arr,
    age_binary_test,
    privileged_value=0  # Under 40 = privileged
)
age_assessment = assess_fairness(age_metrics)
fairness_results['Age'] = {**age_metrics, **age_assessment}
print(f"  Statistical Parity Difference: {age_metrics['statistical_parity_difference']:.4f}")
print(f"  Disparate Impact Ratio: {age_metrics['disparate_impact_ratio']:.4f}")
print(f"  Equal Opportunity Difference: {age_metrics['equal_opportunity_difference']:.4f}")

# 4. DISABILITY FAIRNESS
print("\nCalculating fairness metrics for DISABILITY (Simulated)...")
disability_metrics = calculate_fairness_metrics(
    y_test_arr,
    y_pred_arr,
    X_test['disability'].values,
    privileged_value=0  # Without disability = privileged
)
disability_assessment = assess_fairness(disability_metrics)
fairness_results['Disability'] = {**disability_metrics, **disability_assessment}
print(f"  Statistical Parity Difference: {disability_metrics['statistical_parity_difference']:.4f}")
print(f"  Disparate Impact Ratio: {disability_metrics['disparate_impact_ratio']:.4f}")
print(f"  Equal Opportunity Difference: {disability_metrics['equal_opportunity_difference']:.4f}")

# =============================================================================
# CREATE COMPREHENSIVE RESULTS TABLE
# =============================================================================

display(HTML("""<div class="subsection-header">Fairness Metrics Summary Table</div>"""))

# Build results dataframe
results_data = []
for attr, metrics in fairness_results.items():
    results_data.append({
        'Protected Attribute': attr,
        'SPD': f"{metrics['statistical_parity_difference']:.4f}",
        'SPD Status': metrics['spd_status'],
        'DIR': f"{metrics['disparate_impact_ratio']:.4f}",
        'DIR Status': metrics['dir_status'],
        'EOD': f"{metrics['equal_opportunity_difference']:.4f}",
        'EOD Status': metrics['eod_status'],
        'Overall': 'FAIR' if metrics['overall_fair'] else 'BIASED'
    })

results_df = pd.DataFrame(results_data)

# Create styled HTML table
def get_status_badge(status):
    if status == 'PASS' or status == 'FAIR':
        return f'<span class="badge badge-pass">{status}</span>'
    else:
        return f'<span class="badge badge-fail">{status}</span>'

table_html = """
<table class="styled-table">
    <thead>
        <tr>
            <th>Protected Attribute</th>
            <th>Statistical Parity Diff.</th>
            <th>Status</th>
            <th>Disparate Impact Ratio</th>
            <th>Status</th>
            <th>Equal Opp. Diff.</th>
            <th>Status</th>
            <th>Overall</th>
        </tr>
    </thead>
    <tbody>
"""

for _, row in results_df.iterrows():
    table_html += f"""
        <tr>
            <td><strong>{row['Protected Attribute']}</strong></td>
            <td>{row['SPD']}</td>
            <td>{get_status_badge(row['SPD Status'])}</td>
            <td>{row['DIR']}</td>
            <td>{get_status_badge(row['DIR Status'])}</td>
            <td>{row['EOD']}</td>
            <td>{get_status_badge(row['EOD Status'])}</td>
            <td>{get_status_badge(row['Overall'])}</td>
        </tr>
    """

table_html += "</tbody></table>"

display(HTML(table_html))

# =============================================================================
# INTERPRETATION GUIDE
# =============================================================================

display(HTML("""
<div class="annotation">
    <strong>How to Read This Table:</strong><br><br>
    <strong>SPD (Statistical Parity Difference):</strong> Should be between -0.1 and 0.1. 
    Negative values mean the unprivileged group receives fewer positive outcomes.<br><br>
    <strong>DIR (Disparate Impact Ratio):</strong> Should be between 0.8 and 1.25. 
    Values below 0.8 indicate the unprivileged group is selected at less than 80% the rate of the privileged group.<br><br>
    <strong>EOD (Equal Opportunity Difference):</strong> Should be between -0.1 and 0.1. 
    Negative values mean qualified members of the unprivileged group are less likely to be correctly identified.
</div>
"""))

# Count failures
total_attributes = len(fairness_results)
biased_attributes = sum(1 for r in fairness_results.values() if not r['overall_fair'])

if biased_attributes > 0:
    display(HTML(f"""
    <div class="warning-box">
        <strong>Bias Detected</strong><br><br>
        The baseline model shows bias against <strong>{biased_attributes} out of {total_attributes}</strong> 
        protected attributes. Under South African law (Employment Equity Act Section 6 and Constitution 
        Section 9), this AI system would likely be considered unfairly discriminatory and would require 
        remediation before deployment.
    </div>
    """))
else:
    display(HTML("""
    <div class="success-box">
        <strong>No Significant Bias Detected</strong><br><br>
        The baseline model meets fairness thresholds for all protected attributes.
    </div>
    """))

display(HTML('<div class="section-divider"></div>'))
```

```python
# =============================================================================
# CELL 10: MITIGATION 1 - REWEIGHING (PRE-PROCESSING)
# =============================================================================
# Applying IBM AIF360's Reweighing algorithm to mitigate bias
# =============================================================================

display(HTML("""
<div class="section-header">
    Section 7: Bias Mitigation  Reweighing (Pre-processing)
</div>
"""))

display(HTML("""
<div class="legal-box">
    <div class="legal-box-title">Legal Context: Proactive Measures Under Employment Equity Act</div>
    <p style="margin: 0; color: #374151; line-height: 1.7;">
        The Employment Equity Act requires designated employers to implement <strong>affirmative action 
        measures</strong> to achieve equitable representation. In the AI context, this translates to 
        proactively adjusting systems to ensure fair outcomes. Reweighing is a pre-processing technique 
        that adjusts the training data to remove historical bias before the model learns from it.
    </p>
</div>
"""))

# =============================================================================
# EXPLANATION OF REWEIGHING
# =============================================================================

reweighing_explanation = """
<div style="background: white; border-radius: 16px; padding: 30px; margin: 20px 0;
            box-shadow: 0 4px 20px rgba(0,0,0,0.08); border-left: 6px solid #8B5CF6;">
    <h3 style="color: #8B5CF6; margin: 0 0 20px 0;">How Reweighing Works</h3>
    
    <div style="display: grid; grid-template-columns: 1fr 1fr; gap: 20px;">
        <div>
            <h4 style="color: #6D28D9; margin: 0 0 10px 0;">The Problem</h4>
            <p style="color: #374151; margin: 0; line-height: 1.7;">
                Historical data often reflects past discrimination. For example, if women were 
                historically less likely to be promoted, the training data will show this pattern, 
                and the AI will learn to replicate it.
            </p>
        </div>
        <div>
            <h4 style="color: #6D28D9; margin: 0 0 10px 0;">The Solution</h4>
            <p style="color: #374151; margin: 0; line-height: 1.7;">
                Reweighing assigns weights to training samples to balance the influence of 
                different groups. Underrepresented combinations (e.g., women with positive outcomes) 
                receive higher weights, while overrepresented combinations receive lower weights.
            </p>
        </div>
    </div>
    
    <div style="background: #F9FAFB; padding: 15px 20px; border-radius: 8px; margin-top: 20px;">
        <strong style="color: #6D28D9;">Key Advantage:</strong>
        <span style="color: #374151;"> Reweighing modifies only the training process, not the 
        model architecture. The resulting model can be deployed without any runtime modifications.</span>
    </div>
</div>
"""

display(HTML(reweighing_explanation))

# =============================================================================
# APPLY REWEIGHING FOR GENDER
# =============================================================================

display(HTML("""<div class="subsection-header">Applying Reweighing for Gender Fairness</div>"""))

print("Creating AIF360 datasets for reweighing...")

# Prepare training data for AIF360
train_df_aif = X_train.copy()
train_df_aif['target'] = y_train.values

# Create BinaryLabelDataset
dataset_train = BinaryLabelDataset(
    df=train_df_aif,
    label_names=['target'],
    protected_attribute_names=['gender'],
    favorable_label=1,
    unfavorable_label=0
)

print(f"Original dataset shape: {dataset_train.features.shape}")

# Apply Reweighing
print("\nApplying Reweighing algorithm...")

reweighing = Reweighing(
    unprivileged_groups=[{'gender': 0}],  # Female
    privileged_groups=[{'gender': 1}]      # Male
)

dataset_train_reweighed = reweighing.fit_transform(dataset_train)

# Extract weights
sample_weights = dataset_train_reweighed.instance_weights

print(f"Weights computed for {len(sample_weights):,} samples")
print(f"Weight range: {sample_weights.min():.4f} to {sample_weights.max():.4f}")
print(f"Mean weight: {sample_weights.mean():.4f}")

# Analyze weight distribution
weight_df = pd.DataFrame({
    'Gender': ['Female' if g == 0 else 'Male' for g in X_train['gender']],
    'Original Label': y_train.values,
    'Weight': sample_weights
})

weight_summary = weight_df.groupby(['Gender', 'Original Label'])['Weight'].agg(['mean', 'count'])
print("\nWeight distribution by group:")
print(weight_summary)

# =============================================================================
# TRAIN MODEL WITH REWEIGHED DATA
# =============================================================================

display(HTML("""<div class="subsection-header">Training Reweighed Model</div>"""))

print("Training Logistic Regression with sample weights...")

reweighed_model = LogisticRegression(
    max_iter=1000,
    random_state=42,
    solver='lbfgs'
)

# Train with weights
reweighed_model.fit(X_train_scaled, y_train, sample_weight=sample_weights)

# Predictions
y_pred_reweighed = reweighed_model.predict(X_test_scaled)
y_prob_reweighed = reweighed_model.predict_proba(X_test_scaled)[:, 1]

print("Model training complete!")

# =============================================================================
# EVALUATE REWEIGHED MODEL
# =============================================================================

display(HTML("""<div class="subsection-header">Reweighed Model Evaluation</div>"""))

# Performance metrics
rw_accuracy = accuracy_score(y_test, y_pred_reweighed)
rw_precision = precision_score(y_test, y_pred_reweighed)
rw_recall = recall_score(y_test, y_pred_reweighed)
rw_f1 = f1_score(y_test, y_pred_reweighed)
rw_auc = roc_auc_score(y_test, y_prob_reweighed)

print("=" * 60)
print("REWEIGHED MODEL PERFORMANCE")
print("=" * 60)
print(f"Accuracy:   {rw_accuracy:.4f} (Baseline: {test_accuracy:.4f})")
print(f"Precision:  {rw_precision:.4f} (Baseline: {test_precision:.4f})")
print(f"Recall:     {rw_recall:.4f} (Baseline: {test_recall:.4f})")
print(f"F1 Score:   {rw_f1:.4f} (Baseline: {test_f1:.4f})")
print(f"AUC-ROC:    {rw_auc:.4f} (Baseline: {test_auc:.4f})")
print("=" * 60)

# =============================================================================
# CALCULATE FAIRNESS METRICS AFTER REWEIGHING
# =============================================================================

display(HTML("""<div class="subsection-header">Fairness Metrics After Reweighing</div>"""))

# Calculate fairness metrics for reweighed model
reweighed_fairness = {}

# Gender
gender_metrics_rw = calculate_fairness_metrics(
    y_test.values, y_pred_reweighed, X_test['gender'].values, privileged_value=1
)
gender_assessment_rw = assess_fairness(gender_metrics_rw)
reweighed_fairness['Gender'] = {**gender_metrics_rw, **gender_assessment_rw}

# Race
race_metrics_rw = calculate_fairness_metrics(
    y_test.values, y_pred_reweighed, X_test['race_binary'].values, privileged_value=1
)
race_assessment_rw = assess_fairness(race_metrics_rw)
reweighed_fairness['Race'] = {**race_metrics_rw, **race_assessment_rw}

# Age
age_metrics_rw = calculate_fairness_metrics(
    y_test.values, y_pred_reweighed, age_binary_test, privileged_value=0
)
age_assessment_rw = assess_fairness(age_metrics_rw)
reweighed_fairness['Age'] = {**age_metrics_rw, **age_assessment_rw}

# Disability
disability_metrics_rw = calculate_fairness_metrics(
    y_test.values, y_pred_reweighed, X_test['disability'].values, privileged_value=0
)
disability_assessment_rw = assess_fairness(disability_metrics_rw)
reweighed_fairness['Disability'] = {**disability_metrics_rw, **disability_assessment_rw}

# Create comparison table
print("\nFairness Metrics Comparison (Baseline vs Reweighed):")
print("-" * 80)
print(f"{'Attribute':<12} {'Metric':<8} {'Baseline':>12} {'Reweighed':>12} {'Change':>12} {'Status':>10}")
print("-" * 80)

for attr in ['Gender', 'Race', 'Age', 'Disability']:
    base = fairness_results[attr]
    rw = reweighed_fairness[attr]
    
    # SPD
    spd_change = rw['statistical_parity_difference'] - base['statistical_parity_difference']
    print(f"{attr:<12} {'SPD':<8} {base['statistical_parity_difference']:>12.4f} {rw['statistical_parity_difference']:>12.4f} {spd_change:>+12.4f} {rw['spd_status']:>10}")
    
    # DIR
    dir_change = rw['disparate_impact_ratio'] - base['disparate_impact_ratio']
    print(f"{'':<12} {'DIR':<8} {base['disparate_impact_ratio']:>12.4f} {rw['disparate_impact_ratio']:>12.4f} {dir_change:>+12.4f} {rw['dir_status']:>10}")
    
    # EOD
    eod_change = rw['equal_opportunity_difference'] - base['equal_opportunity_difference']
    print(f"{'':<12} {'EOD':<8} {base['equal_opportunity_difference']:>12.4f} {rw['equal_opportunity_difference']:>12.4f} {eod_change:>+12.4f} {rw['eod_status']:>10}")
    print("-" * 80)

# Store reweighed predictions for later comparison
reweighed_predictions = y_pred_reweighed.copy()
reweighed_probabilities = y_prob_reweighed.copy()

display(HTML("""
<div class="success-box">
    <strong>Reweighing Applied Successfully</strong><br><br>
    The reweighing pre-processing technique has been applied. The comparison above shows how 
    fairness metrics changed after adjusting the training data weights. Improvements toward 
    0 for SPD and EOD, and toward 1.0 for DIR, indicate bias reduction.
</div>
"""))

display(HTML('<div class="section-divider"></div>'))
```

```python
# =============================================================================
# CELL 11: MITIGATION 2 - REJECT OPTION CLASSIFICATION (POST-PROCESSING)
# =============================================================================
# Applying AIF360's Reject Option Classification to mitigate bias
# =============================================================================

display(HTML("""
<div class="section-header">
    Section 8: Bias Mitigation  Reject Option Classification (Post-processing)
</div>
"""))

display(HTML("""
<div class="legal-box">
    <div class="legal-box-title">Legal Context: Section 71 of POPIA  Human Oversight</div>
    <p style="margin: 0; color: #374151; line-height: 1.7;">
        POPIA Section 71 requires that individuals not be subject to decisions based 
        <strong>solely</strong> on automated processing. Reject Option Classification aligns with 
        this principle by identifying cases where the model is uncertain and potentially adjusting 
        outcomes in favor of disadvantaged groupsmimicking the human review process.
    </p>
</div>
"""))

# =============================================================================
# EXPLANATION OF REJECT OPTION CLASSIFICATION
# =============================================================================

roc_explanation = """
<div style="background: white; border-radius: 16px; padding: 30px; margin: 20px 0;
            box-shadow: 0 4px 20px rgba(0,0,0,0.08); border-left: 6px solid #F472B6;">
    <h3 style="color: #F472B6; margin: 0 0 20px 0;">How Reject Option Classification Works</h3>
    
    <div style="display: grid; grid-template-columns: 1fr 1fr; gap: 20px;">
        <div>
            <h4 style="color: #6D28D9; margin: 0 0 10px 0;">The Concept</h4>
            <p style="color: #374151; margin: 0; line-height: 1.7;">
                When a classifier's prediction probability is close to the decision threshold 
                (e.g., between 0.4 and 0.6 for a 0.5 threshold), the prediction is uncertain. 
                ROC identifies these "critical region" cases.
            </p>
        </div>
        <div>
            <h4 style="color: #6D28D9; margin: 0 0 10px 0;">The Adjustment</h4>
            <p style="color: #374151; margin: 0; line-height: 1.7;">
                For uncertain predictions, ROC changes unfavorable outcomes for unprivileged 
                groups to favorable, and vice versa for privileged groups. This promotes equity 
                without changing high-confidence predictions.
            </p>
        </div>
    </div>
    
    <div style="background: #F9FAFB; padding: 15px 20px; border-radius: 8px; margin-top: 20px;">
        <strong style="color: #6D28D9;">Key Advantage:</strong>
        <span style="color: #374151;"> ROC is a post-processing techniqueit works on an already-trained 
        model's outputs. This means it can be applied to any existing AI system without retraining.</span>
    </div>
</div>
"""

display(HTML(roc_explanation))

# =============================================================================
# PREPARE DATA FOR ROC
# =============================================================================

display(HTML("""<div class="subsection-header">Applying Reject Option Classification</div>"""))

print("Preparing datasets for Reject Option Classification...")

# Create AIF360 datasets with predictions
# We need to create a dataset that includes the model's predictions

# Test dataset for AIF360
test_df_aif = X_test.copy()
test_df_aif['target'] = y_test.values

dataset_test_orig = BinaryLabelDataset(
    df=test_df_aif,
    label_names=['target'],
    protected_attribute_names=['gender'],
    favorable_label=1,
    unfavorable_label=0
)

# Create predicted dataset (copy structure, replace labels with predictions)
dataset_test_pred = dataset_test_orig.copy()
dataset_test_pred.labels = y_pred_test.reshape(-1, 1)
dataset_test_pred.scores = y_prob_test.reshape(-1, 1)

print(f"Test dataset prepared: {dataset_test_pred.features.shape[0]} samples")

# =============================================================================
# APPLY REJECT OPTION CLASSIFICATION
# =============================================================================

print("\nApplying Reject Option Classification...")

# Define ROC parameters
roc = RejectOptionClassification(
    unprivileged_groups=[{'gender': 0}],  # Female
    privileged_groups=[{'gender': 1}],     # Male
    low_class_thresh=0.3,                  # Lower bound of critical region
    high_class_thresh=0.7,                 # Upper bound of critical region
    num_class_thresh=100,                  # Number of thresholds to try
    num_ROC_margin=50,                     # Number of margins to try
    metric_name='Statistical parity difference',
    metric_ub=0.05,                        # Upper bound for metric
    metric_lb=-0.05                        # Lower bound for metric
)

# Fit and transform
roc_model = roc.fit(dataset_test_orig, dataset_test_pred)
dataset_test_roc = roc_model.predict(dataset_test_pred)

# Extract ROC predictions
y_pred_roc = dataset_test_roc.labels.flatten().astype(int)

print(f"ROC threshold: {roc_model.classification_threshold:.4f}")
print(f"ROC margin: {roc_model.ROC_margin:.4f}")

# =============================================================================
# EVALUATE ROC MODEL
# =============================================================================

display(HTML("""<div class="subsection-header">Reject Option Classification Evaluation</div>"""))

# Performance metrics
roc_accuracy = accuracy_score(y_test, y_pred_roc)
roc_precision = precision_score(y_test, y_pred_roc)
roc_recall = recall_score(y_test, y_pred_roc)
roc_f1 = f1_score(y_test, y_pred_roc)

print("=" * 60)
print("REJECT OPTION CLASSIFICATION PERFORMANCE")
print("=" * 60)
print(f"Accuracy:   {roc_accuracy:.4f} (Baseline: {test_accuracy:.4f})")
print(f"Precision:  {roc_precision:.4f} (Baseline: {test_precision:.4f})")
print(f"Recall:     {roc_recall:.4f} (Baseline: {test_recall:.4f})")
print(f"F1 Score:   {roc_f1:.4f} (Baseline: {test_f1:.4f})")
print("=" * 60)

# Count changed predictions
predictions_changed = (y_pred_test != y_pred_roc).sum()
print(f"\nPredictions modified by ROC: {predictions_changed:,} ({predictions_changed/len(y_pred_test)*100:.2f}%)")

# =============================================================================
# CALCULATE FAIRNESS METRICS AFTER ROC
# =============================================================================

display(HTML("""<div class="subsection-header">Fairness Metrics After Reject Option Classification</div>"""))

# Calculate fairness metrics
roc_fairness = {}

# Gender
gender_metrics_roc = calculate_fairness_metrics(
    y_test.values, y_pred_roc, X_test['gender'].values, privileged_value=1
)
gender_assessment_roc = assess_fairness(gender_metrics_roc)
roc_fairness['Gender'] = {**gender_metrics_roc, **gender_assessment_roc}

# Race
race_metrics_roc = calculate_fairness_metrics(
    y_test.values, y_pred_roc, X_test['race_binary'].values, privileged_value=1
)
race_assessment_roc = assess_fairness(race_metrics_roc)
roc_fairness['Race'] = {**race_metrics_roc, **race_assessment_roc}

# Age
age_metrics_roc = calculate_fairness_metrics(
    y_test.values, y_pred_roc, age_binary_test, privileged_value=0
)
age_assessment_roc = assess_fairness(age_metrics_roc)
roc_fairness['Age'] = {**age_metrics_roc, **age_assessment_roc}

# Disability
disability_metrics_roc = calculate_fairness_metrics(
    y_test.values, y_pred_roc, X_test['disability'].values, privileged_value=0
)
disability_assessment_roc = assess_fairness(disability_metrics_roc)
roc_fairness['Disability'] = {**disability_metrics_roc, **disability_assessment_roc}

# Print comparison
print("\nFairness Metrics Comparison (Baseline vs ROC):")
print("-" * 80)
print(f"{'Attribute':<12} {'Metric':<8} {'Baseline':>12} {'ROC':>12} {'Change':>12} {'Status':>10}")
print("-" * 80)

for attr in ['Gender', 'Race', 'Age', 'Disability']:
    base = fairness_results[attr]
    roc_res = roc_fairness[attr]
    
    # SPD
    spd_change = roc_res['statistical_parity_difference'] - base['statistical_parity_difference']
    print(f"{attr:<12} {'SPD':<8} {base['statistical_parity_difference']:>12.4f} {roc_res['statistical_parity_difference']:>12.4f} {spd_change:>+12.4f} {roc_res['spd_status']:>10}")
    
    # DIR
    dir_change = roc_res['disparate_impact_ratio'] - base['disparate_impact_ratio']
    print(f"{'':<12} {'DIR':<8} {base['disparate_impact_ratio']:>12.4f} {roc_res['disparate_impact_ratio']:>12.4f} {dir_change:>+12.4f} {roc_res['dir_status']:>10}")
    
    # EOD
    eod_change = roc_res['equal_opportunity_difference'] - base['equal_opportunity_difference']
    print(f"{'':<12} {'EOD':<8} {base['equal_opportunity_difference']:>12.4f} {roc_res['equal_opportunity_difference']:>12.4f} {eod_change:>+12.4f} {roc_res['eod_status']:>10}")
    print("-" * 80)

# Store ROC predictions
roc_predictions = y_pred_roc.copy()

display(HTML("""
<div class="success-box">
    <strong>Reject Option Classification Applied Successfully</strong><br><br>
    The ROC post-processing technique has been applied. This method specifically targets 
    uncertain predictions near the decision boundary, adjusting them to improve fairness 
    while minimizing impact on overall accuracy.
</div>
"""))

display(HTML('<div class="section-divider"></div>'))
```

```python
# =============================================================================
# CELL 14: STATISTICAL TESTING
# =============================================================================
# Validating fairness findings with statistical significance tests
# =============================================================================

display(HTML("""
<div class="section-header">
    Section 11: Statistical Validation of Findings
</div>
"""))

display(HTML("""
<div class="annotation">
    <strong>Why Statistical Testing Matters:</strong> Visual differences in charts may or may not 
    be statistically significant. Statistical tests help us determine whether observed disparities 
    are likely real patterns or could have occurred by chance. This is crucial for legal 
    defensibility of audit findings.
</div>
"""))

# =============================================================================
# CHI-SQUARE TESTS FOR INDEPENDENCE
# =============================================================================

display(HTML("""<div class="subsection-header">Chi-Square Tests for Independence</div>"""))

def perform_chi_square_test(data, group_col, outcome_col):
    """
    Perform chi-square test for independence between group membership and outcome.
    
    Returns test statistic, p-value, and interpretation.
    """
    contingency_table = pd.crosstab(data[group_col], data[outcome_col])
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)
    
    return {
        'chi2_statistic': chi2,
        'p_value': p_value,
        'degrees_of_freedom': dof,
        'significant': p_value < 0.05,
        'contingency_table': contingency_table
    }

# Prepare test data
test_data_stats = X_test.copy()
test_data_stats['actual'] = y_test.values
test_data_stats['predicted'] = y_pred_test
test_data_stats['predicted_roc'] = y_pred_roc

# Chi-square tests for each attribute
chi_square_results = {}

# Gender
chi_square_results['Gender'] = perform_chi_square_test(
    test_data_stats, 'gender', 'predicted'
)

# Race
chi_square_results['Race'] = perform_chi_square_test(
    test_data_stats, 'race_binary', 'predicted'
)

# Age
test_data_stats['age_binary'] = (test_data_stats['age'] >= 40).astype(int)
chi_square_results['Age'] = perform_chi_square_test(
    test_data_stats, 'age_binary', 'predicted'
)

# Disability
chi_square_results['Disability'] = perform_chi_square_test(
    test_data_stats, 'disability', 'predicted'
)

# Display results
print("=" * 80)
print("CHI-SQUARE TEST RESULTS: Independence of Predictions and Protected Attributes")
print("=" * 80)

# FIXED: Define header strings outside of f-string to avoid backslash issues
header_attr = "Attribute"
header_chi = "Chi-Square"
header_pval = "p-value"
header_sig = "Significant"
header_interp = "Interpretation"

print(f"{header_attr:<15} {header_chi:>12} {header_pval:>15} {header_sig:>12} {header_interp:<30}")
print("-" * 80)

for attr, results in chi_square_results.items():
    sig_text = "YES" if results['significant'] else "NO"
    interp = "Predictions depend on attribute" if results['significant'] else "No significant dependence"
    print(f"{attr:<15} {results['chi2_statistic']:>12.4f} {results['p_value']:>15.2e} {sig_text:>12} {interp:<30}")

print("-" * 80)
print("\nNote: p-value < 0.05 indicates statistically significant association between")
print("      predictions and protected attribute (potential unfair discrimination)")

# =============================================================================
# MANN-WHITNEY U TESTS FOR PROBABILITY SCORES
# =============================================================================

display(HTML("""<div class="subsection-header">Mann-Whitney U Tests for Prediction Score Distributions</div>"""))

def perform_mann_whitney_test(scores, group_col, group_val_1, group_val_2, group_labels):
    """
    Perform Mann-Whitney U test to compare score distributions between groups.
    """
    group1_scores = scores[group_col == group_val_1]
    group2_scores = scores[group_col == group_val_2]
    
    statistic, p_value = mannwhitneyu(group1_scores, group2_scores, alternative='two-sided')
    
    return {
        'statistic': statistic,
        'p_value': p_value,
        'significant': p_value < 0.05,
        'group1_median': np.median(group1_scores),
        'group2_median': np.median(group2_scores),
        'group1_label': group_labels[0],
        'group2_label': group_labels[1]
    }

# Mann-Whitney tests
mw_results = {}

mw_results['Gender'] = perform_mann_whitney_test(
    y_prob_test, X_test['gender'].values, 1, 0, ['Male', 'Female']
)

mw_results['Race'] = perform_mann_whitney_test(
    y_prob_test, X_test['race_binary'].values, 1, 0, ['White', 'Non-White']
)

mw_results['Age'] = perform_mann_whitney_test(
    y_prob_test, (X_test['age'] >= 40).astype(int).values, 0, 1, ['Under 40', '40+']
)

mw_results['Disability'] = perform_mann_whitney_test(
    y_prob_test, X_test['disability'].values, 0, 1, ['No Disability', 'With Disability']
)

print("\n" + "=" * 90)
print("MANN-WHITNEY U TEST RESULTS: Comparing Prediction Score Distributions")
print("=" * 90)

# FIXED: Define headers outside f-string
mw_headers = ["Attribute", "Group 1", "Median 1", "Group 2", "Median 2", "p-value", "Sig."]
print(f"{mw_headers[0]:<12} {mw_headers[1]:<15} {mw_headers[2]:>10} {mw_headers[3]:<18} {mw_headers[4]:>10} {mw_headers[5]:>12} {mw_headers[6]:>6}")
print("-" * 90)

for attr, results in mw_results.items():
    sig_text = "YES" if results['significant'] else "NO"
    print(f"{attr:<12} {results['group1_label']:<15} {results['group1_median']:>10.4f} "
          f"{results['group2_label']:<18} {results['group2_median']:>10.4f} "
          f"{results['p_value']:>12.2e} {sig_text:>6}")

print("-" * 90)

# =============================================================================
# EFFECT SIZE CALCULATIONS
# =============================================================================

display(HTML("""<div class="subsection-header">Effect Size Analysis (Cohens h)</div>"""))

import math  # Import at the top level, not inside function

def cohens_h(p1, p2):
    """
    Calculate Cohens h effect size for comparing two proportions.
    h = 2 * (arcsin(sqrt(p1)) - arcsin(sqrt(p2)))
    
    Interpretation:
    - Small: |h| = 0.2
    - Medium: |h| = 0.5
    - Large: |h| = 0.8
    """
    # Handle edge cases
    p1 = max(0.0001, min(0.9999, p1))
    p2 = max(0.0001, min(0.9999, p2))
    
    phi1 = 2 * math.asin(math.sqrt(p1))
    phi2 = 2 * math.asin(math.sqrt(p2))
    return phi1 - phi2

def interpret_cohens_h(h):
    """Interpret Cohens h value"""
    h_abs = abs(h)
    if h_abs < 0.2:
        return "Negligible"
    elif h_abs < 0.5:
        return "Small"
    elif h_abs < 0.8:
        return "Medium"
    else:
        return "Large"

# Calculate effect sizes
effect_sizes = {}

for attr in ['Gender', 'Race', 'Age', 'Disability']:
    p_priv = fairness_results[attr]['privileged_positive_rate']
    p_unpriv = fairness_results[attr]['unprivileged_positive_rate']
    
    h = cohens_h(p_priv, p_unpriv)
    effect_sizes[attr] = {
        'cohens_h': h,
        'interpretation': interpret_cohens_h(h),
        'p_privileged': p_priv,
        'p_unprivileged': p_unpriv
    }

print("\n" + "=" * 80)
print("EFFECT SIZE ANALYSIS: Cohens h for Selection Rate Differences")
print("=" * 80)

# FIXED: Define header labels as variables to avoid backslash in f-string
col1 = "Attribute"
col2 = "Priv. Rate"
col3 = "Unpriv. Rate"
col4 = "Cohens h"  # Removed apostrophe to avoid any issues
col5 = "Effect Size"

print(f"{col1:<15} {col2:>12} {col3:>14} {col4:>12} {col5:<15}")
print("-" * 80)

for attr, results in effect_sizes.items():
    print(f"{attr:<15} {results['p_privileged']:>12.4f} {results['p_unprivileged']:>14.4f} "
          f"{results['cohens_h']:>12.4f} {results['interpretation']:<15}")

print("-" * 80)

# =============================================================================
# STATISTICAL SUMMARY VISUALIZATION
# =============================================================================

display(HTML("""<div class="subsection-header">Statistical Summary Visualization</div>"""))

# Create summary visualization
stat_summary_data = pd.DataFrame({
    'Attribute': list(effect_sizes.keys()),
    'Effect Size': [abs(v['cohens_h']) for v in effect_sizes.values()],
    'Chi-Square Significant': [chi_square_results[a]['significant'] for a in effect_sizes.keys()],
    'Mann-Whitney Significant': [mw_results[a]['significant'] for a in effect_sizes.keys()]
})

fig_stats = go.Figure()

# Effect size bars
fig_stats.add_trace(go.Bar(
    x=stat_summary_data['Attribute'],
    y=stat_summary_data['Effect Size'],
    marker_color=[COLORS['purple'], COLORS['pink'], COLORS['teal'], COLORS['dark_purple']],
    text=[f"{v:.3f}" for v in stat_summary_data['Effect Size']],
    textposition='outside',
    name='Cohens h (absolute)'
))

# Add threshold lines
fig_stats.add_hline(y=0.2, line_dash="dash", line_color=COLORS['teal'],
                   annotation_text="Small Effect (0.2)", annotation_position="right")
fig_stats.add_hline(y=0.5, line_dash="dash", line_color=COLORS['pink'],
                   annotation_text="Medium Effect (0.5)", annotation_position="right")
fig_stats.add_hline(y=0.8, line_dash="dash", line_color=COLORS['dark_purple'],
                   annotation_text="Large Effect (0.8)", annotation_position="right")

fig_stats.update_layout(
    title=dict(
        text="Effect Size Summary by Protected Attribute<br><sub>Cohens h measures practical significance of disparities</sub>",
        font=dict(size=18, color=COLORS['dark_purple']),
        x=0.5, xanchor='center'
    ),
    xaxis_title="Protected Attribute",
    yaxis_title="Absolute Effect Size (Cohens h)",
    yaxis=dict(range=[0, 1]),
    paper_bgcolor=COLORS['white'],
    plot_bgcolor=COLORS['light_gray'],
    height=450,
    width=800,
    showlegend=False
)

fig_stats.show()

# =============================================================================
# SIGNIFICANCE SUMMARY TABLE
# =============================================================================

display(HTML("""<div class="subsection-header">Statistical Significance Summary</div>"""))

# Create HTML summary table
sig_table_html = """
<table class="styled-table" style="margin: 20px 0;">
    <thead>
        <tr>
            <th>Protected Attribute</th>
            <th>Chi-Square Test</th>
            <th>Mann-Whitney U Test</th>
            <th>Effect Size (Cohens h)</th>
            <th>Practical Significance</th>
        </tr>
    </thead>
    <tbody>
"""

for attr in ['Gender', 'Race', 'Age', 'Disability']:
    chi_sig = chi_square_results[attr]['significant']
    mw_sig = mw_results[attr]['significant']
    effect = effect_sizes[attr]
    
    chi_badge = '<span class="badge badge-fail">Significant</span>' if chi_sig else '<span class="badge badge-pass">Not Significant</span>'
    mw_badge = '<span class="badge badge-fail">Significant</span>' if mw_sig else '<span class="badge badge-pass">Not Significant</span>'
    
    # Effect size badge
    if abs(effect['cohens_h']) >= 0.8:
        effect_badge = '<span class="badge badge-fail">Large</span>'
    elif abs(effect['cohens_h']) >= 0.5:
        effect_badge = '<span class="badge badge-warning">Medium</span>'
    elif abs(effect['cohens_h']) >= 0.2:
        effect_badge = '<span class="badge badge-warning">Small</span>'
    else:
        effect_badge = '<span class="badge badge-pass">Negligible</span>'
    
    sig_table_html += f"""
        <tr>
            <td><strong>{attr}</strong></td>
            <td style="text-align: center;">{chi_badge}<br><small>p = {chi_square_results[attr]['p_value']:.2e}</small></td>
            <td style="text-align: center;">{mw_badge}<br><small>p = {mw_results[attr]['p_value']:.2e}</small></td>
            <td style="text-align: center;">{effect['cohens_h']:.4f}</td>
            <td style="text-align: center;">{effect_badge}</td>
        </tr>
    """

sig_table_html += """
    </tbody>
</table>
"""

display(HTML(sig_table_html))

# =============================================================================
# INTERPRETATION GUIDE
# =============================================================================

display(HTML("""
<div class="annotation">
    <strong>How to Interpret These Results:</strong><br><br>
    
    <strong>Chi-Square Test:</strong> Tests whether model predictions are independent of 
    group membership. A significant result (p < 0.05) means the model treats groups differently.<br><br>
    
    <strong>Mann-Whitney U Test:</strong> Compares the distribution of prediction scores between 
    groups. Significance indicates systematic differences in how the model scores different groups.<br><br>
    
    <strong>Cohens h Effect Size:</strong> Measures the practical magnitude of the difference:
    <ul>
        <li>< 0.2: Negligible - difference unlikely to matter in practice</li>
        <li>0.2 - 0.5: Small - noticeable but minor impact</li>
        <li>0.5 - 0.8: Medium - substantial practical impact</li>
        <li>> 0.8: Large - major practical significance, requires attention</li>
    </ul>
</div>
"""))

# =============================================================================
# KEY FINDINGS
# =============================================================================

display(HTML("""
<div class="key-finding">
    <div class="key-finding-title">Statistical Validation Summary</div>
    <p style="margin: 0; color: #374151; line-height: 1.8;">
        The statistical tests confirm our visual findings:<br><br>
        
        <strong>Chi-Square Tests:</strong> Significant associations between protected attributes 
        and model predictions indicate the models decisions are not independent of group membership.<br><br>
        
        <strong>Mann-Whitney U Tests:</strong> Significant differences in prediction score 
        distributions confirm systematic scoring differences between groups.<br><br>
        
        <strong>Effect Sizes:</strong> Cohens h values quantify the practical significance 
        of disparities. Values above 0.5 indicate medium to large effects that are likely 
        to have real-world impact on hiring outcomes.<br><br>
        
        <strong>Legal Implication:</strong> These statistically validated disparities provide 
        evidence that would be relevant in an Employment Equity Act or Constitutional challenge. 
        The combination of statistical significance AND practical significance (effect size) 
        strengthens the case for bias remediation.
    </p>
</div>
"""))

display(HTML('<div class="section-divider"></div>'))
```
